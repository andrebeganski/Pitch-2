```python
##Import pandas
import pandas as pd
```

```python
##Basic info for BLS data:
##I have four seperate excel documents from the BLS that I need to combine into one dataset in Excel.

##The goal is to make an index that compares the LFPR recovery of people with and without disabilites.

##The files all measure labor force participation rate (LFPR)
###FILE 1 - LFPR for men without a disability
###FILE 2 - LFPR for women without a disability
###FILE 3 - LFPR for men with a disability
###FILE 4 - LFPR for women with a disability

##I got the data here: https://www.bls.gov/webapps/legacy/cpsatab6.htm
```

```python
##Importing FILE 1 and checking datatypes
df_f1 = pd.read_excel("data/File1.xlsx", skiprows=12)
df_f1.info()
```

```python
##Looking at FILE 1
df_f1
```

```python
##Melting FILE 1
df_f1 = df_f1.melt(id_vars = "Year",
                  var_name = "month",
                  value_name = "LFPR")
df_f1
```

```python
##Adding gender and disability status for FILE 1
df_f1["gender"] = "male"
df_f1["disabled"] = False
df_f1
```

```python
##Importing FILE 2 and looking at it
df_f2 = pd.read_excel("data/File2.xlsx", skiprows=12)
df_f2
```

```python
##Melting FILE 2
df_f2 = df_f2.melt(id_vars = ["Year"],
                  var_name = "month",
                  value_name = "LFPR")
df_f2
```

```python
##Adding gender and disability status for FILE 2
df_f2["gender"] = "female"
df_f2["disabled"] = False
df_f2
```

```python
##Importing FILE 3 and looking at it
df_f3 = pd.read_excel("data/File3.xlsx", skiprows=12)
df_f3
```

```python
##Melting FILE 3
df_f3 = df_f3.melt(id_vars = ["Year"],
                  var_name = "month",
                  value_name = "LFPR")
df_f3
```

```python
##Adding gender and disability status for FILE 3
df_f3["gender"] = "male"
df_f3["disabled"] = True
df_f3
```

```python
##Importing FILE 4 and looking at it
df_f4 = pd.read_excel("data/File4.xlsx", skiprows=12)
df_f4
```

```python
##Melting FILE 4
df_f4 = df_f4.melt(id_vars = ["Year"],
                  var_name = "month",
                  value_name = "LFPR")
df_f4
```

```python
##Adding gender and disability status for FILE 4
df_f4["gender"] = "female"
df_f4["disabled"] = True
df_f4
```

```python
##Combing all files into single dataframe
df_all = pd.concat([df_f1, df_f2, df_f3, df_f4], sort=False)
df_all
```

```python
##Dropping values for months that have not happened yet
df_all = df_all.dropna()
df_all
```

```python
##Changing year from number to date
df_all["year"] = pd.to_datetime(df_all["Year"], format='%Y')
df_all["year_only"] = df_all["year"].dt.year
df_all.drop(["Year", "year"], inplace = True, axis = 1)
df_all
```

```python
##Checking to see if values for 2022 are accurate
mask_1 = df_all["year_only"] == 2022
df_all[mask_1]
```

```python
##Checking datatypes
df_all.info()
```

```python
##Converting abbreviated months into values compatible with datetime format
d = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
df_all["month"] = df_all["month"].map(d)
df_all
```

```python
##Converting abbreviated months into datetime format
df_all["month_only"] = pd.to_datetime(df_all["month"], format='%m')
df_all
```

```python
##Dropping day and year from month datetime format
df_all["month_only"] = df_all["month_only"].dt.month
df_all.drop(["month"], inplace = True, axis = 1)
df_all
```

```python
##Renaming year and month columns to combine into singular datetime
df_all = df_all.rename(columns={"year_only": "YEAR"})
df_all = df_all.rename(columns={"month_only": "MONTH"})
df_all
```

```python
##combiining values into datetime format
df_all['date'] = pd.to_datetime(df_all[[ 'YEAR', 'MONTH']].assign(DAY=1))
df_all
```

```python
##Dropping year and month column that are no longer needed
df_all.drop(["MONTH", "YEAR"], inplace = True, axis = 1)
df_all
```

```python
##Masking data into new dataframe that gives me values for after pandemic began
mask_2 = df_all['date'] > '2020-01-01'
df_post = df_all[mask_2]
df_post
```

```python
##Making mask for people that are male and do not have a disability
mask_3 = (df_post['gender'] == 'male') & (df_post['disabled'] == False)
df_post_m_nd = df_post[mask_3]
df_post_m_nd
```

```python
##Making mask for people that are female and do not have a disability
mask_4 = (df_post['gender'] == 'female') & (df_post['disabled'] == False)
df_post_f_nd = df_post[mask_4]
df_post_f_nd
```

```python
##Creating dataframe for the index to compare labor force participation for people with and without a disability
df_post_comp = df_post_f_nd
df_post_comp.info()
```

```python
##Creating column that is the average of LFPR for people without a disability
df_post_comp['LFPR'] = (df_post_f_nd["LFPR"]+df_post_m_nd["LFPR"])/2
df_post_comp
```

```python
##Renaming column for people without aq disability to make it more accurate
df_post_comp = df_post_comp.rename(columns={"LFPR": "LFPR_not_disabled"})
df_post_comp
```

```python
##Creating mask for women with a disability
mask_5 = (df_post['gender'] == 'female') & (df_post['disabled'] == True)
df_post_f_d = df_post[mask_5]
df_post_f_d
```

```python
##Creating mask for men with a disability
mask_6 = (df_post['gender'] == 'male') & (df_post['disabled'] == True)
df_post_m_d = df_post[mask_6]
df_post_m_d
```

```python
##Creating column that is the average of LFPR for people with a disability
df_post_comp['gender'] = (df_post_f_d["LFPR"]+df_post_m_d["LFPR"])/2
df_post_comp
```

```python
##Dropping column that is no longer relevant
df_post_comp.drop(["disabled"], inplace = True, axis = 1)
df_post_comp
```

```python
##Renaming column that is the average of LFPR for people without a disability to be more accurate
df_post_comp = df_post_comp.rename(columns={"gender": "LFPR_disabled"})
df_post_comp
```

```python
##Sorting values by date
df_post_comp = df_post_comp.sort_values(by="date", ascending = True)
df_post_comp
```

```python
##Creating column for percent change for people without a disability
LFPR_not_disabled_fixed = 77.75
df_post_comp["LFPR_not_disabled_pct_chg"] = df_post_comp.apply \
(lambda row: ((row['LFPR_not_disabled'] - LFPR_not_disabled_fixed)/LFPR_not_disabled_fixed)*100, axis= 1) \
.round(1)
df_post_comp
```

```python
##Creating column for percent change for people with a disability
LFPR_disabled_fixed = 33.80
df_post_comp["LFPR_disabled_pct_chg"] = df_post_comp.apply \
(lambda row: ((row['LFPR_disabled'] - LFPR_disabled_fixed)/LFPR_disabled_fixed)*100, axis= 1) \
.round(1)
df_post_comp
```

```python
##Analysis:
##The labor force participation rate for people without a disability has not fully recovered from Feb of 2020
##The labor force participation rate for people with a disability has grown by 11.7% from Feb of 2020
```
